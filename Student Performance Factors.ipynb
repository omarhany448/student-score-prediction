<h3>Outliers</h3>

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df[['Hours_Studied','Exam_Score','Previous_Scores']])
plt.title("Boxplot to Detect Outliers")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)  
    Q3 = df[col].quantile(0.75)  
    IQR = Q3 - Q1                

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]

    print(f"Column: {col}")
    print(f"Number of outliers: {outliers.shape[0]}")
    if not outliers.empty:
        print(f"Example outliers:\n{outliers.head(10)}")
    print("-"*50)
```

```python
def remove_outliers_iqr(df, columns):
    df_clean = df.copy()
    for col in columns:
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5*IQR
        upper = Q3 + 1.5*IQR
        df_clean = df_clean[(df_clean[col] >= lower) & (df_clean[col] <= upper)]
    return df_clean

numeric_cols = df.select_dtypes(include=['int64','float64']).columns
df_clean = remove_outliers_iqr(df, numeric_cols)

print(f"Original shape: {df.shape}")
print(f"After removing outliers: {df_clean.shape}")
```

```python
X_clean = df_clean.drop(columns=[target])
y_clean = df_clean[target]

X_clean = pd.get_dummies(X_clean, drop_first=True)

Xc_train, Xc_test, yc_train, yc_test = train_test_split(
    X_clean, y_clean, test_size=0.2, random_state=42
)

model_clean = LinearRegression()
model_clean.fit(Xc_train, yc_train)

y_pred_clean = model_clean.predict(Xc_test)

r2_clean = r2_score(yc_test, y_pred_clean)
rmse_clean = np.sqrt(mean_squared_error(yc_test, y_pred_clean))

print("=== Model without Outliers ===")
print(f"R Score: {r2_clean:.3f}")
print(f"RMSE: {rmse_clean:.3f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

X_poly = df_clean.drop(columns=['Exam_Score'])
y_poly = df_clean['Exam_Score']

X_poly = pd.get_dummies(X_poly, drop_first=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error



Xp_train, Xp_test, yp_train, yp_test = train_test_split(
    X_poly, y_poly, test_size=0.2, random_state=42
)

poly_model = Pipeline([
    ('poly_features', PolynomialFeatures(degree=2, include_bias=False)),
    ('lin_reg', LinearRegression())
])

poly_model.fit(Xp_train, yp_train)

y_poly_pred = poly_model.predict(Xp_test)


r2_poly = r2_score(yp_test, y_poly_pred)
rmse_poly = np.sqrt(mean_squared_error(yp_test, y_poly_pred))

print(f"Polynomial Regression R: {r2_poly:.3f}")
print(f"Polynomial Regression RMSE: {rmse_poly:.3f}")
```
